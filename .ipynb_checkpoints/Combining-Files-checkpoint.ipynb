```python
#Import Dependencies

import pandas as pd
import os, fnmatch
# from splinter import Browser
# from bs4 import BeautifulSoup as bs
import datetime as dt
import zipfile
import numpy as np
```

```python
# To unzip all downloaded files and move them into "2015Q4_2020_unzipped"

for root, dirs, files in os.walk("2015Q4_2020", topdown=False):
    for name in files:
        if os.path.join(name).endswith(".zip"):
            x = os.path.join(root, name)
            print(x)
            with zipfile.ZipFile(f"{x}", 'r') as zip_ref:
                zip_ref.extractall("2015Q4_2020_unzipped")
```

```python
# defining file_search(): to use for extracting with respect to each year.

def file_search():
    for root, dirs, files in os.walk("2015Q4_2020_unzipped", topdown=False):
            # Creating lists to store files after looping
            Y2015Files = []
            Y2016Files = []
            Y2017Files = []
            Y2018Files = []
            Y2019Files = []
            Y2020Files = []
            Other = []
            AllFiles = []
            # Run through the whole folder to look and categorise yearly and also to an AllFiles folder. 
            for name in files:
                if os.path.join(name).startswith("2015", 3, 7):
                    Y2015Files.append(os.path.join(root, name))
                    AllFiles.append(os.path.join(root, name))
                elif os.path.join(name).startswith("2016", 3, 7):
                    Y2016Files.append(os.path.join(root, name))
                    AllFiles.append(os.path.join(root, name))
                elif os.path.join(name).startswith("2017", 3, 7):
                    Y2017Files.append(os.path.join(root, name))
                    AllFiles.append(os.path.join(root, name))
                elif os.path.join(name).startswith("2018", 3, 7):
                    Y2018Files.append(os.path.join(root, name))
                    AllFiles.append(os.path.join(root, name))
                elif os.path.join(name).startswith("2019", 3, 7):
                    Y2019Files.append(os.path.join(root, name))
                    AllFiles.append(os.path.join(root, name))
                elif os.path.join(name).startswith("2020", 3, 7):
                    Y2020Files.append(os.path.join(root, name))
                    AllFiles.append(os.path.join(root, name))
                else:
                    Other.append(os.path.join(root, name))

#     print(Y2015Files)
#     print(Y2016Files)
#     print(Y2017Files)
#     print(Y2018Files)
#     print(Y2019Files)
#     print(Y2020Files)
#     print(AllFiles)

    # Resources helps to load files of any year with the following inputs
    Resources = {
        "2015": Y2015Files,
        "2016": Y2016Files,
        "2017": Y2017Files,
        "2018": Y2018Files,
        "2019": Y2019Files,
        "2020": Y2020Files,
        "Other": Other,
        "AllFiles": AllFiles
    }
    return Resources
```
